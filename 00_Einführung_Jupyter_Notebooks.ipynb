## 1. Was ist Jupyter Notebooks und Binder

Im Informatik-Unterricht an der Kantonsschule Olten erlernen wir unter anderem den Umgang mit der Programmiersprache Python. Wie jede andere Programmiersprache braucht auch Python eine *Ausführumgebung*. Die Ausführumgebung ist dafür verantwortlich, den von dir geschriebenen Code in computerlesbare Befehle zu "übersetzen" und somit ausführbar zu machen.

Bis jetzt hast du womöglich bereits mit verschiedenen Ausführumgebungen gearbeitet, eine davon ist *TigerJython*. TigerJython kann Python-Befehle ausführen, aus didaktischen Gründen jedoch nur eingeschränkt. TigerJython kann entweder Text oder das Bild einer Turtle-Zeichnung zurückgeben.

**Jupyter Notebooks** ist eine weitere Ausführumgebung für Python. Jupyter Notebooks bestehen aus zwei Teilen:

- Einem **Notebook**. Das Notebook ist eine Datei mit der Endung .ipynb. In das Notebook schreibst du deinen Code.
- Einem **Jupyter Server**. Dieser Server kann mit einer Web-Adresse aufgerufen werden und führt dein Notebook im Browser aus. Wir verwenden dafür **Binder**.

Jupyter Notebooks verfügen über sogenannte Zellen (engl. Cell). Eine Zelle kann entweder Text enthalten (so wie diese hier) oder ausführbaren Python-Code. Wird eine Text-Zelle ausgeführt, dann wird lediglich der Text angezeigt. Wir eine Code-Zelle ausgeführt, dann wird das Ergebnis des prozessierten Codes angezeigt.

## 2. Jupyter Notebooks auf Binder hoch und runterladen

Binder ist eine hervorragende Ausführumgebung für Jupyter Notebooks, sie hat allerdings einen Haken: **Nach 20 Minuten Inaktivität wird Binder heruntergefahren und nichts wird gespeichert.**

***
>Wir müssen also jedes Mal, wenn wir Binder verlassen unser Notebook (die Datei mit der Endung .ipynb) auf unseren Computer herunterladen und dort speichern. Wenn wir wieder damit arbeiten wollen, müssen wir das Notebook wieder hochladen.
***

### Code-Zelle ausführen
> Um eine Code-Zelle auszuführen musst du sie angewählt haben und im Anschluss auf den **Run-Knopf** in der Menü-Zeile klicken oder die Tastenkombination **ctrl + Enter** klicken.

### Zelle erstellen
> Um eine neue Zelle zu erstellen klickst du entweder auf den **+-Knopf** oder die Tastenkombination **Esc + b**. Im folgenden wird eine neue Zelle unterhalt der angewählten Zelle erstellt.

### Notebook neu starten

Es kann passieren, dass bei langem Bearbeiten der Speicher des Notebooks ein wenig durcheinander kommt. In diesem Fall kannst du das Notebook mit einem Klick auf das Refreshsymbol (runder Pfeil) zurücksetzen. All dein Text und Code bleibt erhalten, aber die Zellen müssen neu ausgeführt werden. Mit einem Klick auf den Spuhl-Knopf (Doppelpfeil) wird dein Notebook zurückgesetzt und alle Zellen automatisch neu ausgeführt.
